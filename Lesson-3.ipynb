## Conditionals

if elif else when etc etc


```python
x = 1
if x > 0:
    print("x is positive")
    print("This is another line of code inside the if block")
else:
    print("x is not positive")
```

## Chained Conditions

```python
x = 1
y = -1

if x>y:
    print("x is greater than y")
elif x<y:
    print("x is less than y")
else:
    print("x is equal to y")
```

```python
if x == y:
    print("x is equal to y")
else:
    if x > y:
        print("x is greater than y")
    else:
        print("x is less than y")
```

```python
if (x > 0 and y < 0) or  (x < 0 and y > 0):
    print("x is positive and y is negative")
else:
    print("Either x is not positive or y is not negative")
```

## Exception Catching

```python
x = 100
y = 0

try:
    x = int(x)
    y = int(y)
    sweets_per_student = x // y
    print(f"Each student gets {sweets_per_student} sweets.")
except ValueError as e:
    print("Invalid input. Please enter integers for sweets and students.")
except Exception as trump:
    print("I am here")
    print(trump)
```
